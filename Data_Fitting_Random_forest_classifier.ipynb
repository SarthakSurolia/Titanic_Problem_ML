```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split as tts
from sklearn.ensemble import RandomForestClassifier as rfc
```

```python
train = pd.read_csv(r"C:\Users\hp\Desktop\titanic problem\New folder\Notebook\clean_train.csv")
test = pd.read_csv(r"C:\Users\hp\Desktop\titanic problem\New folder\Notebook\clean_test.csv")
```

```python
train
```

### We have to drop PassengerId column from both our sets.

```python
train = train.drop(["PassengerId"], axis = 1)
id_ = test["PassengerId"] 
test = test.drop(["PassengerId"], axis = 1)

Y = train["Survived"]
train = train.drop(["Survived"], axis = 1)
test = test.drop(["Survived"], axis = 1)
X = train.copy()
X_test = test.copy()
```

```python
X
```

### Now we are going to use Random Forest Classifier as our model to fit the training set. Before that we split our Training set into train and cross validation set in order to prevent overfitting of our model and to check the performance of our model. Also for Random Forest Classifiers it is not required to do feature scaling or normalization of our Parameters

```python
X_train, X_cv, Y_train, Y_cv = tts(X, Y, test_size = 0.25, random_state = 0)
```

### We use scikit learn library to implement random forest classifer. For this classifier we can play with many parameters such as max_depth, max_features, n_estimators, etc. And sometimes default values for them is not fit for our model. So in order to find perfect combination of parameters, we use Hyperparameter tuning. We will use GridSearchCV module for it. 
### Also we will use a 3 fold crossvalidation method.

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
rf_cls = rfc()
grid_search = GridSearchCV(estimator = rf_cls, param_grid = param_grid, 
                           cv = 3, n_jobs = -1, verbose = 2)
grid_search.fit(X_train, Y_train)
```

```python
grid_search.best_params_
```

```python
pred_cv = grid_search.predict(X_cv)
from sklearn.metrics import classification_report
print(classification_report(Y_cv, pred_cv))
```

### Our F1 score for cross validation set is 0.87 for y = 0 and 0.74 for y = 1
### Now we also check the f1 score of our training set in order to make sure if it is overfitting or not.

```python
pred = grid_search.predict(X_train)
print(classification_report(Y_train, pred))
```

### So it is seen that Our Train and CV F1 Scores are similar. This means our model did not actually overfit the training set.
### We fit our model on training set and then convert it into desired format required to submit our Predictions.

```python
pred_test = grid_search.predict(X_test)

res = pd.DataFrame(data= [id_, pred_test], index= ["PassengerId", "Survived"])
result_ = res.transpose()
result_.to_csv(r"C:\Users\hp\Desktop\titanic problem\New folder\Notebook\result_rf.csv", index= False)
```
